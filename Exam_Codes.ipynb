````python
from flask import Flask, request, jsonify, render_template
import joblib
import pandas as pd
import numpy as np
import os

MODEL_PATH = "gbr_model.joblib"
FEATURES = [
    'energy', 'duration', 'acousticness', 'danceability', 'tempo', 'speechiness',
    'key', 'liveness', 'time_signature', 'mode', 'loudness', 'valence',
    'Song.Quality', 'Is.Final', 'Year'
]

app = Flask(__name__, template_folder="templates")

if not os.path.exists(MODEL_PATH):
    raise FileNotFoundError(f"Model file '{MODEL_PATH}' not found. Run save_model.py to create it.")

saved = joblib.load(MODEL_PATH)
if isinstance(saved, dict) and "model" in saved:
    model = saved['model']
    meta = saved.get('meta', {})
    FEATURES = meta.get('features', FEATURES)
    TRAIN_MEDIANS = meta.get('medians', {})
else:
    model = saved
    TRAIN_MEDIANS = {}

@app.route("/", methods=["GET"])
def index():
    return render_template("index.html")

def build_dataframe_from_input(input_dicts):
    X = pd.DataFrame(input_dicts)
    for feat in FEATURES:
        if feat not in X.columns:
            X[feat] = np.nan
    X = X[FEATURES].copy()
    for col in X.columns:
        X[col] = pd.to_numeric(X[col], errors='coerce')
    fill_vals = {col: TRAIN_MEDIANS.get(col, 0.0) for col in FEATURES}
    X = X.fillna(pd.Series(fill_vals))
    return X

@app.route("/predict", methods=["POST"])
def predict():
    json_data = None
    try:
        json_data = request.get_json(silent=True)
    except Exception:
        json_data = None

    instances = None
    single_input = False

    if json_data:
        if isinstance(json_data, dict) and "instances" in json_data and isinstance(json_data["instances"], list):
            instances = json_data["instances"]
        elif isinstance(json_data, list):
            instances = json_data
        elif isinstance(json_data, dict):
            instances = [json_data]
        else:
            return jsonify({"error": "Invalid JSON format."}), 400
    else:
        form = request.form
        if not form:
            return jsonify({"error": "No JSON body and no form data found."}), 400
        single_input = True
        inst = {}
        for feat in FEATURES:
            if feat == 'Is.Final':
                inst[feat] = form.get(feat, '0')
            else:
                inst[feat] = form.get(feat, form.get(feat.lower(), np.nan))
        instances = [inst]

    X = build_dataframe_from_input(instances)
    try:
        preds = model.predict(X)
    except Exception as e:
        return jsonify({"error": f"Prediction failed: {str(e)}"}), 500

    preds_list = [float(x) for x in preds]

    if single_input:
        prediction_text = f"Predicted points: {preds_list[0]:.2f}"
        return render_template("index.html", prediction_text=prediction_text)
    else:
        return jsonify({"predictions": preds_list}), 200

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True)
````

````python
import os
import pandas as pd
import joblib
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor

EXCEL_PATH = "partpdf_1763620554909_eurovision_1998 to 2012.xlsx"
MODEL_PATH = "gbr_model.joblib"

features = [
    'energy', 'duration', 'acousticness', 'danceability', 'tempo', 'speechiness',
    'key', 'liveness', 'time_signature', 'mode', 'loudness', 'valence',
    'Song.Quality', 'Is.Final', 'Year'
]

if not os.path.exists(EXCEL_PATH):
    raise FileNotFoundError(f"Excel file not found at {EXCEL_PATH}. Place your dataset there or update EXCEL_PATH.")

df = pd.read_excel(EXCEL_PATH)

if 'Points' not in df.columns:
    raise KeyError("Target column 'Points' not found in Excel file.")

df = df[features + ['Points']].copy()
df = df.dropna(subset=['Points'])
train_medians = {}
for col in features:
    if col in df.columns:
        if not pd.api.types.is_numeric_dtype(df[col]):
            df[col] = pd.to_numeric(df[col], errors='coerce')
        med = df[col].median()
        if pd.isna(med):
            med = 0.0
        train_medians[col] = float(med)
        df[col] = df[col].fillna(med)
    else:
        df[col] = 0.0
        train_medians[col] = 0.0

X = df[features]
y = df['Points']
model = GradientBoostingRegressor(random_state=42)
model.fit(X, y)
saved = {"model": model, "meta": {"features": features, "medians": train_medians}}
joblib.dump(saved, MODEL_PATH)
print(f"Trained GradientBoostingRegressor on {len(X)} rows and saved model+meta to '{MODEL_PATH}'.")
````
